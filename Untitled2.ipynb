```python
import geopandas as gpd
from shapely.geometry import Polygon
import rasterio
from rasterio.mask import mask
import cv2
import numpy as np
import requests
from io import BytesIO
import os
```

```python

image_path = 'SURINOVA_CBE_DTM.tif'
img = rasterio.open(image_path)

def classify_vegetation(image):
    threshold = 0.5  # Adjust as needed
    vegetation_mask = image.read(1) > threshold
    return vegetation_mask

def calculate_area(polygons):

    return polygons.area

def width_analysis(vegetation_mask):
    kernel = np.ones((1, 10), np.uint8)
    width_mask = cv2.morphologyEx(vegetation_mask.astype(np.uint8), cv2.MORPH_CLOSE, kernel)
    return width_mask


```

```python
# Add print statements to identify potential issues
print("Performing vegetation classification...")
vegetation_mask = classify_vegetation(img)
print(vegetation_mask)
```

```python

print("Creating GeoDataFrame with vegetation polygons...")
contours, _ = cv2.findContours(vegetation_mask.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
polygons = [Polygon(cv2.approxPolyDP(cnt, 0.01, True).reshape(-1, 2)) for cnt in contours]
gdf = gpd.GeoDataFrame(geometry=polygons, crs=img.crs)
gdf['area'] = gdf.area
print('\n')
print("Generated GeoDataFrame:\n", gdf.head())
```

```python

output_directory = 'output'

os.makedirs(output_directory, exist_ok=True)

print("Saving GeoDataFrame to shapefile...")
gdf.to_file(os.path.join(output_directory, 'vegetation_polygons.shp'))
```

```python
print("Saving area estimation to CSV...")
gdf[['geometry', 'area']].to_csv(os.path.join(output_directory, 'area_estimation.csv'), index=False)
# print(area)

# Perform width analysis
```

```python
# Perform width analysis
print("Performing width analysis...")
width_mask = width_analysis(vegetation_mask)
print(width_mask)
```

```python
# Save width analysis result
print("Saving width analysis result to PNG...")
cv2.imwrite(os.path.join(output_directory, 'width_analysis.png'), (width_mask * 255).astype(np.uint8))
```

```python
# Output data for navigation
print("Calculating navigation data...")
navigation_data = gdf.representative_point()
navigation_data = gpd.GeoDataFrame(geometry=navigation_data, crs=gdf.crs)
# print(navigation_data)
```

```python
# Save navigation data to GeoJSON
print("Saving navigation data to GeoJSON...")
navigation_data.to_file(os.path.join(output_directory, 'navigation_map.geojson'), driver='GeoJSON')
print("Generated navigation data:\n", navigation_data.head())
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(12, 12))

ax.imshow(img.read(1), cmap='gray', alpha=0.5)

ax.imshow(vegetation_mask, cmap='viridis', alpha=0.3)
for geom in gdf.geometry:
    if geom.geom_type == 'Polygon':

        patch = patches.Polygon(np.array(geom.exterior.xy).T, edgecolor='red', facecolor='none', linewidth=2)
        ax.add_patch(patch)
    elif geom.geom_type == 'MultiPolygon':

        for polygon in geom:
            patch = patches.Polygon(np.array(polygon.exterior.xy).T, edgecolor='red', facecolor='none', linewidth=2)
            ax.add_patch(patch)

ax.imshow(width_mask, cmap='Blues', alpha=0.3)

ax.set_title('Vegetation Classification, Vegetation Polygons, and Width Analysis')
ax.set_axis_off()
output_image_path = os.path.join(output_directory, 'composite_image.png')
plt.savefig(output_image_path, bbox_inches='tight', pad_inches=0.1, dpi=300)
plt.show()
```
